```python
# Part 1: Starting Simple - Create Initial Prompts
# Objective: Create basic zero-shot prompts for three different task types and establish a baseline.

# Step 1: Setup Your Environment
# What to do:

# Create a new Jupyter notebook or Python script named
# prompt_engineering_lab.ipynb
# Set up your OpenAI client and create helper functions for testing (running multiple calls to OpenAI at the same time)
# Expected outcome: You have a working environment with helper functions ready to use.

# Checkpoint: Run the setup code and verify you get the success message.


import os
from openai import OpenAI
import json
from collections import Counter
import time
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Initialize OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

#set default model
MODEL = "gpt-4o-mini"

print("Environment setup complete. Ready to create prompts!")
```

```python
def run_openai_request(prompt, model=MODEL, temperature=0.1, max_tokens=600):
    """
    Function to run a request to the OpenAI API with the given prompt and model.
    Returns the response text.
    """
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            temperature=temperature,
            max_tokens=max_tokens,
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error during OpenAI request: {e}")
        return None
```

```python
def run_multiple_requests(prompt,n,temp):
    responses_list = []
    for i in range(n):
        response = run_openai_request(prompt, temperature=temp)
        responses_list.append(response)
        # print(f"Response {i+1}: {response}")
        time.sleep(1)  # Sleep for 1 second between requests to avoid rate limits
    return responses_list
```

## Failure Analysis & Consistency Report

To quantify the improvement from v1 → v2 → v3, every prompt version is run **15 times** and scored on two metrics:

- **Exact-match consistency** — the share of runs that produced the single most common response. Strict, and the right metric for classification (sentiment) and extraction, where a production system needs byte-identical, parseable output.
- **Format adherence** — the share of runs that satisfy the task's format rules (single-line label, ~50-word plain prose, all required field names). This is the fair metric for the generative product-description task, where wording is *expected* to vary but structure must not. The format rules are the **final production spec**, applied uniformly to all versions, so earlier versions are scored against the same target the final prompt must hit.

Every version is tested at the **same temperature (1.0)**, so the only variable changing between versions is the prompt itself — any consistency gain is attributable purely to prompt design, not to temperature tuning.

```python
# Failure analysis helpers
# Two metrics per prompt:
#  - exact-match consistency: % of runs equal to the most common response (strict; suits classification/extraction)
#  - format adherence: % of runs that satisfy the task's format rules (suits generative tasks where wording may vary)

import re

def word_count(text):
    return len(text.split())

def sentiment_format_ok(r):
    # Expected: one line ending with a single allowed label word
    return r is not None and len(r.strip().splitlines()) == 1 and \
        bool(re.search(r'\b(Positive|Negative|Neutral)\.?$', r.strip()))

def product_format_ok(r):
    # Expected: ~50 words (40-70 tolerance), plain prose without markdown headers/bullets
    return r is not None and 40 <= word_count(r) <= 70 and "**" not in r and "\n-" not in r

def extraction_format_ok(r):
    # Expected: all four required field names present, exactly as specified
    required = ["Item ID", "Order Date", "Delivery Status", "Packaging Condition"]
    return r is not None and all(field in r for field in required)

def analyze_consistency(responses, format_checker):
    counts = Counter(responses)
    most_common_resp, most_common_n = counts.most_common(1)[0]
    n = len(responses)
    return {
        "runs": n,
        "unique_responses": len(counts),
        "exact_match_consistency_pct": round(most_common_n / n * 100, 1),
        "format_adherence_pct": round(sum(bool(format_checker(r)) for r in responses) / n * 100, 1),
        "most_common": most_common_resp,
    }

print("Analysis helpers ready.")
```

```python
# Collect 15-run results for every task x version, all at the same temperature (1) for a fair comparison.
# This gives one comparable dataset for the failure analysis table.

test_matrix = {
    ("Sentiment", "v1"): (sentiment_prompt_v1, 1),
    ("Sentiment", "v2"): (sentiment_prompt_v2, 1),
    ("Sentiment", "v3"): (sentiment_prompt_v3, 1),
    ("Product", "v1"): (product_prompt_v1, 1),
    ("Product", "v2"): (product_prompt_v2, 1),
    ("Product", "v3"): (product_prompt_v3, 1),
    ("Extraction", "v1"): (extraction_prompt_v1, 1),
    ("Extraction", "v2"): (extraction_prompt_v2, 1),
    ("Extraction", "v3"): (extraction_prompt_v3, 1),
}

all_results = {}
for (task, version), (prompt, temp) in test_matrix.items():
    print(f"Running {task} {version} - 15 runs at temperature {temp}...")
    all_results[(task, version)] = run_multiple_requests(prompt, 15, temp)

print("Done collecting results.")
```

```python
# Failure Analysis Report: consistency table across all tasks and versions

format_checkers = {
    "Sentiment": sentiment_format_ok,
    "Product": product_format_ok,
    "Extraction": extraction_format_ok,
}

report = {}
print("| Task       | Version | Exact-match consistency | Format adherence | Unique responses (of 15) |")
print("|------------|---------|-------------------------|------------------|--------------------------|")
for (task, version), responses in all_results.items():
    stats = analyze_consistency(responses, format_checkers[task])
    report[(task, version)] = stats
    print(f"| {task:<10} | {version:<7} | {stats['exact_match_consistency_pct']:>22}% | {stats['format_adherence_pct']:>15}% | {stats['unique_responses']:>24} |")

print("\n--- v1 vs v3 improvement ---")
for task in ["Sentiment", "Product", "Extraction"]:
    v1, v3 = report[(task, "v1")], report[(task, "v3")]
    print(f"{task}: exact-match {v1['exact_match_consistency_pct']}% -> {v3['exact_match_consistency_pct']}% | "
          f"format adherence {v1['format_adherence_pct']}% -> {v3['format_adherence_pct']}% | "
          f"unique responses {v1['unique_responses']} -> {v3['unique_responses']}")
```

### Documented Failure Patterns (from the 5/10/15-run tests, all at temperature 1)

**Task 1 — Sentiment Analysis** (exact-match: v1 13.3% → v2 100% → v3 100%)
- **v1 failures:** Never returned a single clean label (11 unique responses in 15 runs). Responses were full sentences ("This customer message can be classified as ..."), often proposed *two* invented categories at once ("Positive Feedback" or "Customer Satisfaction"), and the markdown emphasis, quoting style, and wording changed run to run. A downstream check like `if sentiment == "positive"` would fail on every run.
- **v2 fix:** Constraining the output to a single word from an allowed set was enough for 100% exact-match consistency — even at temperature 1.
- **v3 fix:** Few-shot examples additionally lock the output *shape* (message + " - " + label), keeping 100% consistency while making the format explicit rather than implied.

**Task 2 — Product Description** (format adherence: v1 0% → v2 0% → v3 100%)
- **v1 failures:** Wildly variable length (200–400+ words), invented product names (SwiftConnect, Aero Glide) and invented specs (DPI settings, 33-ft range, AA batteries) that differ per run, and heavy markdown structure with headers/bullets that changed every run.
- **v2 fix:** A 50–100-word limit, tone requirement, and fixed feature list stopped the hallucinated specs — but v2 scores 0% here because the format checker enforces the *final production spec* (~50 words, 40–70 tolerance) uniformly across versions, and v2's looser limit produced ~90-word outputs.
- **v3 fix:** A worked example (few-shot) + a required feature *order* pinned length and structure to the target spec: 100% format adherence. Exact-match stays at 6.7% by nature — a generative task at temperature 1 is *supposed* to vary in wording — which is why format adherence is the right metric for this task.

**Task 3 — Data Extraction** (exact-match: v1 20% → v2 93.3% → v3 6.7%; format adherence: 0% → 100% → 100%)
- **v1 failures:** The model invented its own schema each run (12 unique responses) — field names varied ("Order Number" / "Item Number" / "Order ID"; "Delivery Speed" / "Delivery Experience" / "Issue"), preamble sentences appeared in some runs and not others, and values were unnormalized ("March 15th", "Damaged packaging" vs "Damaged").
- **v2 fix:** An explicit field list with exact names and value formats (DD/MM/YYYY, controlled vocabularies) → 93.3% exact-match, the best strict consistency of any prompt at temperature 1.
- **v3 trade-off (an honest finding):** Adding Chain-of-Thought kept format adherence at 100% — the final structured block is always present and parseable — but *destroyed* exact-match consistency (6.7%), because the model prints its step-by-step reasoning before the result and that reasoning text varies freely between runs. Lesson: CoT improves reasoning reliability but adds variable text; in production you would ask the model to reason *silently* or parse only the final block. Both v2 and v3 also share a residual bug: they confidently output the year "2023" even though the source text contains no year (fixed in Part 5 with a "Not Mentioned" rule).

```python
# Variation 1: Sentiment analysis on different message types (template based on sentiment_prompt_v3)
# The template turns the v3 prompt into a reusable function and adds a Neutral label,
# since real customer messages are not always clearly positive or negative.

def make_sentiment_prompt(message):
    return f"""
Classify this customer message as Positive, Negative, or Neutral: "{message}"
Format: the message in quotes, followed by " - " and exactly one word: Positive, Negative, or Neutral
Example: "I love this product!" - Positive
Example: "This product is terrible." - Negative
Example: "The mouse arrived on Tuesday." - Neutral
"""

sentiment_variations = [
    "The delivery took two weeks and the box was crushed. Very disappointed.",  # clearly negative
    "The mouse works as described.",                                            # neutral / factual
    "Great mouse, but the scroll wheel already squeaks.",                       # mixed sentiment
]

for msg in sentiment_variations:
    print("MESSAGE:", msg)
    for r in run_multiple_requests(make_sentiment_prompt(msg), 5, 1):
        print("   ", r)
    print()
```

```python
# Variation 2: Product descriptions for different products (template based on product_prompt_v3)

def make_product_prompt(product, price, features):
    return f"""
Create a product description for {product} that costs {price}.
The description should be around 50 words and use a professional tone.
Focus only on the product's features and describe them in the following order ({features}), avoiding any personal opinions or unrelated information.
For example:
    Introducing our Wireless Mouse, designed with an ergonomic shape for comfortable use during long hours.
Enjoy extended productivity with a long battery life, ensuring uninterrupted sessions.
The 2.4GHz reliable connectivity offers a seamless experience without lag.
Priced at just $29.99, this mouse delivers exceptional value for money,
combining quality and performance.
"""

product_variations = [
    ("bluetooth headphones", "$49.99", "active noise cancellation, 30-hour battery life, Bluetooth 5.3 connectivity, great value for money"),
    ("a mechanical keyboard", "$89.99", "tactile switches, RGB backlighting, detachable USB-C cable, great value for money"),
]

for product, price, features in product_variations:
    print("PRODUCT:", product, price)
    for r in run_multiple_requests(make_product_prompt(product, price, features), 5, 1):
        print(r)
        print("   ---")
    print()
```

```python
# Variation 3: Data extraction on different feedback texts (template based on extraction_prompt_v3)
# Includes a text with MISSING fields to test how the prompt copes; "Not Mentioned" is allowed
# so the model does not invent values (the year-guessing problem seen in v2/v3).

def make_extraction_prompt(feedback):
    return f"""
Extract the following information from this customer feedback: "{feedback}"
First identify each field in the text step by step,
then output the final structured result in the following order.
If a field is not mentioned in the text, write "Not Mentioned" instead of guessing.
1. Item ID: followed by 5 digits in the format #xxxxx
2. Order Date: followed by the date in the format DD/MM/YYYY (use "Not Mentioned" if the year or date is absent)
3. Delivery Status: followed by either word Delivered/Not Delivered
4. Packaging Condition: followed by the word Good/Damaged
"""

extraction_variations = [
    "I ordered item #98765 on April 2nd 2024. It still has not arrived and customer support was unhelpful.",
    "Item #54321 arrived in perfect packaging. Great service!",  # no date at all
]

for feedback in extraction_variations:
    print("FEEDBACK:", feedback)
    for r in run_multiple_requests(make_extraction_prompt(feedback), 5, 1):
        print(r)
        print("   ---")
    print()
```

### Step 12 findings — what works and what needs adjustment

- **Sentiment template:** generalizes perfectly to clearly negative messages (5/5 "Negative") and holds its format on every variation. Two real weaknesses surfaced: the factual, opinion-free message ("The mouse works as described.") was classified **Positive 5/5 despite the Neutral option** — the model reads "works as described" as satisfaction — and the mixed message ("Great mouse, but the scroll wheel already squeaks.") split 4× Negative / 1× Neutral. Single-label classification hides nuance; a production version would need better Neutral few-shot examples closer to real cases, and possibly a "Mixed" label.
- **Product template:** turning v3 into a function with `product / price / features` parameters worked with no other changes — structure, ~50-word length, and feature order carried over to headphones and a keyboard. The few-shot example does the heavy lifting.
- **Extraction template:** the fixed schema held on new feedback texts, and the added rule to output **"Not Mentioned"** for absent fields worked exactly as intended (5/5 runs on the no-date feedback said "Order Date: Not Mentioned" instead of inventing a year — fixing the hallucinated "2023" seen in v2/v3). The CoT reasoning remains verbose and varies between runs, confirming the Step 10 finding: parse the final block, or ask for silent reasoning.
